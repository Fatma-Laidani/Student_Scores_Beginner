## **Dataset**


```python
import pandas as pd
df = pd.read_csv('/content/drive/MyDrive/mydata/student_scores.csv')
df
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
X = df.drop(['Scores'],axis =1,inplace=False)
X.head(5)
```

```python
y=df['Scores']
y.head()
```

```python
import plotly.express as px
import matplotlib # mathematique plot library
import matplotlib.pyplot as plt # python plotting
import seaborn as sns
%matplotlib inline
```

```python
sns.set_style('darkgrid')
matplotlib.rcParams['font.size']=14
matplotlib.rcParams['figure.figsize']=(10,6)
matplotlib.rcParams['figure.facecolor']='BE7B72'
```

```python
fig=px.histogram(df,
                x='Hours',
                title='Distrubution of Hours')
fig.update_layout(bargap=0.2)
fig.show()
```

```python
fig = px.scatter(df,
                 x = 'Hours',
                 y='Scores',
                 color='Scores',
                 opacity=0.4,
                )
fig.update_traces(marker_size=5)
fig.show()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=.25,random_state=44,shuffle=True)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
lr= LinearRegression(fit_intercept=True,n_jobs=-1)
```

```python
lr.fit(X_train,y_train)
```

```python
y_pred=lr.predict(X_test)
y_pred
```

```python
list(y_pred)
```

```python
df1=pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
df1
```

## **Evalution** with MSE

```python
from sklearn.metrics import mean_squared_error
MSEValue = mean_squared_error(y_test,y_pred,multioutput='uniform_average')
MSEValue
```
